```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('immersed-membrane-bioreactor-evaluation.csv')
print(df)
```

```python
X=df.drop(['Effluent BOD5'],axis=1).values
y=df['Effluent BOD5'].values
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=233)
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(X_train,y_train)
```

```python
print("coeff",mlr.coef_)
```

```python
print('intercept',mlr.intercept_)
```

```python
df.head(10)
```

```python
y_pred=mlr.predict(X_test)
print(y_pred)
```

```python
mlr.predict([[425,3.71,18.33,7.0,280,0.3,4.9,7.0,12000.0,0.21,37.1,2.00]])
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('Y Test')
plt.ylabel('Predicted Values')
```
